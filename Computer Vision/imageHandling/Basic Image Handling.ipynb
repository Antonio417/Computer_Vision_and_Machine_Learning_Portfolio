```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./sampleImage.png')
```

```python
print("The image size is: ",img.shape)
print("Images are typically stored as 8 bit unsigned integers. Type: ",img.dtype)
print("This means an image comprises values between 0 and 255.")
```

```python
# invert color channel
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

# Image Cropping

```python
plt.imshow(img)
plt.plot([100,300],[400,400],'k--') # Top line
plt.plot([100,300],[700,700],'k--') # Bottom line
plt.plot([100,100],[400,700],'k--') # Left line
plt.plot([300,300],[400,700],'k--') # Right line
plt.show()
```

```python
# Take a slice of the image and assign it to a new variable, reorder the channels at the same time
img_crop = img[400:700,100:300]
plt.imshow(img_crop)
plt.show()
```

```python
# Since our image is just a grid of numbers, we can perform mathematical operations on the whole image
mu_red = np.mean(img_crop[:,:,0])
mu_green = np.mean(img_crop[:,:,1])
mu_blue = np.mean(img_crop[:,:,2])
print("Average pixel intensity in each channel: ",mu_red,mu_green,mu_blue)

sd_red = np.std(img_crop[:,:,0])
sd_green = np.std(img_crop[:,:,1])
sd_blue = np.std(img_crop[:,:,2])
print("Standard deviation of pixel intensity in each channel: ",sd_red,sd_green,sd_blue)
```

# Normalizing and Average the Image Channels

```python
img_op = 1/3*((img_crop[:,:,0]-mu_red)/sd_red + (img_crop[:,:,1]-mu_green)/sd_green + (img_crop[:,:,2]-mu_blue)/sd_blue)

plt.imshow(img_op,cmap='gray')
plt.colorbar()
plt.show()
```
