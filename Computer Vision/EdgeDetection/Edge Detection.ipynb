```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

# Load Image

```python
img_path = os.path.join('RunningStockPhoto.jpeg')
# Read Image
img = cv2.imread(img_path)
# Image After Recoloring
img_recolor = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_recolor)
```

# Apply Canny Edge Detection

```python
canny = cv2.Canny(img_recolor, threshold1=180, threshold2=300)
plt.imshow(canny)
```

# Apply Gaussian Blur

```python
blr = cv2.GaussianBlur(img, (5,5), 0)
canny = cv2.Canny(blr, threshold1=180, threshold2=300)
plt.imshow(canny)
```

# Apply Blur and Greyscale

```python
gry = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blr = cv2.GaussianBlur(gry, (5,5), 0)
canny = cv2.Canny(blr, threshold1=210, threshold2=300)
plt.imshow(canny)
```

# Exporting Image

```python
cv2.imwrite('CannyEdgeDetection.jpg', canny)
```
